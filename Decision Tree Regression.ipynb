```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading dataset

```python
Movie=pd.read_csv('E:\SaiVenkat\Movie_regression.csv')
Movie
```

```python
Movie.info() #checking whether there are any null values present in dataset
```

```python
Movie['Time_taken'].mean() #as there are null values in time taken variable replacing those with necessary values
```

```python
Movie['Time_taken'].fillna(Movie['Time_taken'].mean(),inplace=True)
```

```python
Movie.info() #checking found no null values
```

```python
Movie['3D_available'].value_counts()
```

```python
Movie['Genre'].value_counts()
```

```python
#assigning dummy values to 3d and genre variable
```

```python
Movie=pd.get_dummies(Movie,columns=['3D_available','Genre'],drop_first=True)
```

```python
#with the help of drop first it will eliminate least occuring value
```

```python
Movie.info()
```

```python
Movie
```

```python
#spliting x and y variables
```

```python
features=Movie.iloc[:,Movie.columns!='Collection']
```

```python
features
```

```python
label=Movie['Collection']
label
```

# splitting data into train and test data set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,label,test_size=0.2,random_state=0)
```

```python
x_train.shape ,y_train.shape , x_test.shape ,y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
regtree=DecisionTreeRegressor(max_depth=3)
regtree.fit(x_train,y_train)
```

# Predicting values

```python
y_train_pred=regtree.predict(x_train)
y_train_pred
```

```python
y_test_pred=regtree.predict(x_test)
y_test_pred
```

# Model evaluation

```python
from sklearn.metrics import mean_squared_error,r2_score
mean_squared_error(y_train,y_train_pred)
```

```python
mean_squared_error(y_test,y_test_pred)
```

```python
r2_score(y_train,y_train_pred)
```

```python
r2_score(y_test,y_test_pred)
```

# controlling the tree growth

```python
#minimum observations at internal node
```

```python
regtree2=DecisionTreeRegressor(min_samples_split=40) 
regtree2.fit(x_train,y_train)
```

```python
dot_data=tree.export_graphviz(regtree2,out_file=None,feature_names=x_train.columns,filled=True)
from IPython.display import Image
import pydotplus
graph2=pydotplus.graph_from_dot_data(dot_data)
Image(graph2.create_png())
```

```python
regtree3=tree.DecisionTreeRegressor(min_samples_leaf=25,max_depth=4)
regtree3.fit(x_train,y_train)
dot_data=tree.export_graphviz(regtree3,out_file=None,feature_names=x_train.columns,filled=True)
from IPython.display import Image
import pydotplus
graph3=pydotplus.graph_from_dot_data(dot_data)
Image(graph3.create_png())
```
